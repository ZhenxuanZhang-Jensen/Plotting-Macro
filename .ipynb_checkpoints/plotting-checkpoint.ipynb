```python
import ROOT
import uproot
import vector
vector.register_awkward()
import pandas as pd
# import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
import awkward as ak
import hist
from hist import Hist,axis
```

```python
events_300 = uproot.open("/eos/user/z/zhenxuan/DoubleHiggs/MC_Root/GluGluToRadionToHHTo2G4Q_M300.root:Events")
```

```python
events_300.show()
```

```python
q1 = events_300["GEN_Q1_pT"].array()
q2 = events_300["GEN_Q2_pT"].array()
q3 = events_300["GEN_Q3_pT"].array()
q4 = events_300["GEN_Q4_pT"].array()
df_orderWinvM = pd.DataFrame()
pt1=[]
pt2=[]
pt3= []
pt4 = []
for i in range(len(q1)):
    pt1.append(q1[i])
    pt2.append(q2[i])
    pt3.append(q3[i])
    pt4.append(q4[i])
df_orderWinvM["pT1"] = pt1
df_orderWinvM["pT2"] = pt2
df_orderWinvM["pT3"] = pt3
df_orderWinvM["pT4"] = pt4
```

```python
df_orderWinvM
```

```python
order_pt1=[]
order_pt2=[]
order_pt3= []
order_pt4 = []
df_orderpT = pd.DataFrame()
for i in range(len(q1)):
    n = sorted(df_orderWinvM.loc[i,:].tolist())
    order_pt1.append(n[3])
    order_pt2.append(n[2])
    order_pt3.append(n[1])
    order_pt4.append(n[0])
    
df_orderpT["pT1"] = order_pt1
df_orderpT["pT2"] = order_pt2
df_orderpT["pT3"] = order_pt3
df_orderpT["pT4"] = order_pt4
```

```python
# len(df_orderpT.iloc[:,3][
[df_orderpT.iloc[:,3] == df_orderWinvM.iloc[:,3]] or [df_orderpT.iloc[:,3] == df_orderWinvM.iloc[:,2]]
```

```python
minx = 0
maxx =100
pick_color = list(mcolors.TABLEAU_COLORS)

for i in range(0,4):
    plt.hist(df_orderpT.iloc[:,i].tolist(),
             bins=100,
                range=(minx,maxx),
                label=df_orderpT.columns[i],
                fill=False,
                linewidth=2,
                edgecolor=pick_color[i],
                histtype='step'
                )
# plt.xlabel("mass")
plt.xlabel(r"$pT(GeV)$")
plt.ylabel("number of events")
plt.legend(loc=1,fontsize=7)
# plt.legend(fontsize=7,bbox_to_anchor=(1,1))
```

```python
pick_color = list(mcolors.TABLEAU_COLORS)
# var_list = ['GEN_Q1_pT','GEN_Q2_pT','GEN_Q3_pT','GEN_Q4_pT']
var_list = ["GEN_W1_Q1Q2_mass","GEN_W2_Q3Q4_mass"]
# var_list = ["GEN_W1_mass","GEN_W2_mass"]
minx = 0
maxx = 100

for i in range(len(var_list)):
    plt.hist(events_300[var_list[i]].array(),
             bins=100,
                range=(minx,maxx),
                label=var_list[i],
                fill=False,
                linewidth=2,
                edgecolor=pick_color[i],
                histtype='step'
                )
# plt.xlabel("mass")
plt.xlabel(r"$pT(GeV)$")
plt.ylabel("number of events")
plt.legend(loc=1,fontsize=7)
# plt.legend(fontsize=7,bbox_to_anchor=(1,1))
```

```python
var1 = "GEN_W1_Q1Q2_mass"
var2 = "GEN_W2_Q3Q4_mass"
minx=miny = 0
maxx = maxy = 100
h = Hist(
    axis.Regular(
        100, minx, maxx, name="A", label=var1, underflow=False, overflow=False
    ),
    axis.Regular(
        100, miny, maxy, name="B", label=var2, underflow=False, overflow=False
    ),
).fill(events_300[var1], events_300[var2])

# plot2d_full
plt.figure(figsize=(8, 8))

h.plot2d_full(
    main_cmap="coolwarm",
    top_ls="--",
    top_color="orange",
    top_lw=2,
    side_ls=":",
    side_lw=2,
    side_color="steelblue",
)

plt.show()
```
